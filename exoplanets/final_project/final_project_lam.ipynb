```python
#x = len(has_planets)
#n = len(hasnt_planets)+len(has_planets)
#print(x, n)

def f(a, b, metallicity):
    """
    takes single values of alpha, beta, and metallicity
    returns single value for probability of having planet
    """
    #print(a, b, metallicity)
    return a*10**(b*metallicity)

def calc_likelihood(metallicity, has_planet_flag, a, b):
    """
    takes DF columns for metallicity and has_planet status, plus alpha and beta values for f()
    returns a single likelihood value
    """
    likelihood = np.zeros(len(metallicity))
    has_planet_flag = list(has_planet_flag)
    for i, m in enumerate(metallicity):      
        # decide whether f or 1-f
        if has_planet_flag[i]=='P':
            likelihood[i] = f(a, b, m)
        elif has_planet_flag[i]=='S':
            likelihood[i] = 1-(f(a, b, m))
    
    return likelihood, np.prod(likelihood)

def apply_prior(likelihood):
    # takes in array and does element-wise application of The Hammer
    for i in range(len(likelihood)):
        if (likelihood[i] <= 0.) or (likelihood[i] > 1.): # if out of bounds of U~[0,1] prior
            likelihood[i] = 0
        else:
            pass # effectively multiplying by 1
    return likelihood

@np.vectorize # for meshgrid version, not used but see junkyard
def calc_likelihood_vectorized(metallicity, has_planet, a, b):
    """
    takes DF columns for metallicity and has_planet status, plus alpha and beta values for f()
    returns a single likelihood value
    """
    likelihood = np.zeros(len(has_label))
        
    # first apply the likelihood itself
    # decide whether f or 1-f
    if has_planet=='P':
        likelihood_elt = f(a, b, metallicity)
    elif has_planet=='S':
        likelihood_elt = 1-(f(a, b, metallicity))
    print(likelihood_elt)
    # then apply the prior
    if (likelihood_elt <= 0) or (likelihood_elt > 1): # if out of bounds of U~[0,1] prior
        pass
    else:
        return likelihood_elt
```
